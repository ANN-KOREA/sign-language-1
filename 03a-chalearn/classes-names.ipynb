```python
import os

import numpy as np
import pandas as pd
```

```python
f = open("../datasets/04-chalearn/class_raw.txt", "r")
sLabels01 = f.read()
f.close()
```

```python
sLabels02 = sLabels01.replace("\n", " ").replace(",", "")
liLabels = sLabels02.split(" ")
```

```python
#liLabels
```

```python
dfLabels = pd.DataFrame({"sClass" : []})
```

```python
nCount = 0
for i in range(len(liLabels)):
    s = liLabels[i]
    if s.isdigit(): 
        nClass = int(s)
        sClass = "c{:03d}".format(nClass)
        sName = liLabels[i+1]
        
        liName = sName.split("/")
        if len(liName) == 1:
            sCat, sDetail = "", sName
        else:
            sCat, sDetail = liName[0], liName[-1]
        print(i, sClass, sName, sCat, sDetail)
        
        dfLabels.loc[nCount, "sClass"] = sClass
        dfLabels.loc[nCount, "sLong"] = sName
        dfLabels.loc[nCount, "sCat"] = sCat
        dfLabels.loc[nCount, "sDetail"] = sDetail
        nCount += 1
        
```
