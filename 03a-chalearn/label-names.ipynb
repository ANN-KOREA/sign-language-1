```python
import os

import numpy as np
import pandas as pd
```

```python
f = open("labels_temp.txt", "r")
sLabels01 = f.read()
f.close()
```

```python
sLabels02 = sLabels01.replace("\n", " ").replace(",", "")
liLabels = sLabels02.split(" ")
```

```python
liLabels
```

```python
dfLabels = pd.DataFrame({"sLabel" : []})
```

```python
nCount = 0
for i in range(len(liLabels)):
    s = liLabels[i]
    if s.isdigit(): 
        nLabel = int(s)
        sLabel = "{:03d}".format(nLabel)
        sName = liLabels[i+1]
        
        liName = sName.split("/")
        if len(liName) == 1:
            sCat, sDetail = "", sName
        else:
            sCat, sDetail = liName[0], liName[-1]
        print(i, sLabel, sName, sCat, sDetail)
        
        dfLabels.loc[nCount, "sLabel"] = sLabel
        dfLabels.loc[nCount, "sNameLong"] = sName
        dfLabels.loc[nCount, "sCat"] = sCat
        dfLabels.loc[nCount, "sDetail"] = sDetail
        nCount += 1
        
```
